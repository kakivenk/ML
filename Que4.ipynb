```python
# Install necessary libraries
!pip install pandas sklearn

# Import libraries
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from google.colab import files

# Upload the CSV file
uploaded = files.upload()

# Load the CSV file into a DataFrame
df = pd.read_csv(next(iter(uploaded.keys())))

# Convert the 'quality' attribute to binary values
df['quality'] = df['quality'].apply(lambda x: 1 if x > 6 else 0)

# Exclude the 'quality' column for normalization
features = df.drop(columns=['quality'])

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the features
normalized_features = scaler.fit_transform(features)

# Create a DataFrame with the normalized features
normalized_df = pd.DataFrame(normalized_features, columns=features.columns)

# Add the 'quality' column back to the normalized DataFrame
normalized_df['quality'] = df['quality'].values

# Save the processed data to a new CSV file
normalized_df.to_csv('normalized_wine_data.csv', index=False)

# Download the processed file
files.download('normalized_wine_data.csv')
```
